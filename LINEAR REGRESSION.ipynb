## LINEAR REGRESSION

#### Linear regression function

```python
import pandas as pd
df = pd.read_csv('salary_data.csv')
df.head(5)
```

```python
df['Salary'].iloc[1]
```

```python
def linear_regression (m_now,b_now,l,data): # m = slope, b = intercept, l = learning rate for 1 independent variable
    m_gradient = 0 # 0 initial
    b_gradient = 0 # 0 initial
    col = list(data.columns)
    n = len(data)
    
    for i in range(len(data)):
        
        y =  data[col[1]].iloc[i] # iterating through rows
        x =  data[col[0]].iloc[i] # iterating through rows
        
        m_gradient += -x * (y - (m_now * x + b_now)) #adding all derivative
        b_gradient += -(y - (m_now * x + b_now))  #adding all derivative
        
    m = m_now - l * (2/n) * m_gradient
    b = b_now - l * (2/n) * b_gradient
    return(m,b)
```

```python
m = 0
b = 0
learn = 0.0001
epoch = 1000
```

```python
a = list(range(0,15))
c = [m*x + b for x in a]
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['YearsExperience'],df['Salary'])
plt.plot(a,c)
plt.show()
```

```python
for i in range(epoch):
    m,b = linear_regression(m,b,learn,df)
    print (m,b)
print ('--------------')
print (m,b)
```
